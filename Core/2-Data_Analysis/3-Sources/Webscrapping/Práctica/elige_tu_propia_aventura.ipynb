* Parsea mediante BeatifulSoup el html y obtén un DataFrame con la información que deseabas.

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
```

```python
url = "https://www.filmaffinity.com/es/ranking.php?rn=ranking_fa_movies"
response = requests.get(url)
```

```python
soup = bs(response.content, 'html.parser')
```

```python
soup
```

```python
titulos = soup.find_all('div', class_ = 'mc-title')
titulos
```

```python
precios = soup.find_all('span', class_ = "_cDEzb_p13n-sc-price_3mJ9Z")

precios
```

```python
ratings = soup.find_all('div', class_ = 'avg-rating')
ratings
```

```python
years = soup.find_all('span', class_ = 'mc-year')
years
```

```python
authors = soup.find_all('div', class_ = "mc-director")
authors
```

```python
peliculas_top = pd.DataFrame({'Título' : [titulo.find('a').get_text() for titulo in titulos[:20]],
                              'Director' : [author.find('a').get_text() for author in authors[:20]],
                              'Año' : [year.get_text() for year in years[:20]],
                                'Rating' : [rating.get_text() for rating in ratings[:20]]})

peliculas_top.index = peliculas_top.index + 1
peliculas_top
```

```python
peliculas_top.to_csv('peliculas_top.csv', encoding='utf-8', index=True)
```
