17. Con el diagrama de violín vemos la forma de la distribución, pero no sabemos si Grass tiene muchos puntos y Fire tiene muy pocos.
Para ver esto, usamos el swarmplot.
Pinta el último diagrama de violines pero con un swarmplot en lugar de un violin

```python
plt.subplot(1,2,1)
sns.swarmplot(data = df_gfw, x = 'Type 1', y = 'Attack', hue = 'Type 1', palette = color_list)
plt.title('Attack')
plt.ylabel('')

plt.subplot(1,2,2)
sns.swarmplot(data = df_gfw, x = 'Type 1', y = 'Defense', color = 'Brown', hue = 'Type 1', palette = color_list)
plt.title('Defense')
plt.ylabel('')
```

 18. Queremos ver los violines a la vez que los swarm, los podemos pintar juntos. 
 Elimina los diagramas de caja dentro de los violines con el parámetro inner = None y pon una transparencia de 0.9 en el swarmplot
 Pon un título con pyplot (como siempre)

```python
plt.subplot(1,2,1)
sns.violinplot(data = df_gfw, x = 'Type 1', y = 'Attack', hue = 'Type 1', palette = color_list, inner = None)
sns.swarmplot(data = df_gfw, x = 'Type 1', y = 'Attack', color = 'Black', alpha = 0.9)
plt.title('Attack')
plt.ylabel('')

plt.subplot(1,2,2)
sns.violinplot(data = df_gfw, x = 'Type 1', y = 'Defense', hue = 'Type 1', palette = color_list, inner = None)
sns.swarmplot(data = df_gfw, x = 'Type 1', y = 'Defense', color = 'Black', alpha = 0.9)
plt.title('Defense')
plt.ylabel('')
```
